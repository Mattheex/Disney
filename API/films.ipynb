```python
import requests
from io import StringIO
import json
```

```python
url = "http://api.disneyapi.dev/character?page=149&pageSize=50"

reponse = requests.get(url)
content = reponse.content
content
```

```python
y = json.loads(content)
```

```python
next= y["info"]["nextPage"]
print(next)
```

```python
films = []
persos = y["data"]
for perso in persos:
    for k in perso["films"]:
        films.append(k)

films
```

```python
len(films)
```

```python
url = "https://api.disneyapi.dev/character"
films=[]
while url:
    reponse = requests.get(url)
    content = reponse.content
    y = json.loads(content)
    url = y["info"]["nextPage"]
    persos = y["data"]
    for perso in persos:
        for k in perso["films"]:
            films.append(k)
```

```python
len(films)
```

```python
filmsUnique = list(set(films))
len(filmsUnique)
```

```python
for k in filmsUnique:
    print(k)
```

```python
import pandas as pd
df = pd.DataFrame(filmsUnique)
df
```
